```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl  

mpl.rcParams['font.sans-serif'] = ['SimHei'] 

P = np.array([[0.6, 0.1, 0.2, 0.1],
              [0.2, 0.5, 0.2, 0.1],
              [0.15,0.05,0.7, 0.1],
              [0.1, 0.1,  0.05,  0.75]])
G = np.array([0.3,0.4,0.2,0.1])

j= int(input("请输入第几年："))

i =1
a=np.array([0.3])
b=np.array([0.4])
c=np.array([0.2])
d=np.array([0.1])

while i<=j:
    G = np.dot(G,P)
    print(G) 
    a=np.append(a,G[0])
    #print(a)
    b=np.append(b,G[1])
    #print(b)
    c=np.append(c,G[2])
    #print(c)
    d=np.append(d,G[3])
    #print(d)
    i += 1
    
print(a)
print(b)
print(c)
print(d)

plt.figure()    
plt.plot(a, label='甲')
plt.plot(b,label= '乙')
plt.plot(c,label= '丙')
plt.plot(d,label= '丁')
plt.legend()
plt.xlabel("年份")
plt.ylabel("占有率")
plt.title("市场占有率预测")
plt.show()
```
