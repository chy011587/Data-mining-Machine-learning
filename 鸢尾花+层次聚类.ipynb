```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
import mglearn
# 从SciPy中导入dendrogram函数和ward聚类函数
from scipy.cluster.hierarchy import dendrogram,single,average,complete
```

```python
iris = load_iris()
X = iris.data[:]
# 将聚类应用于数据数组X
# SciPy的函数返回一个数组，指定执行凝聚聚类时跨越的距离
linkage_array = single(X)
# 现在为包含簇之间距离的linkage_array绘制树状图
dendrogram(linkage_array)

# 在树中标记划分成两个簇或三个簇的位置
ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds, [1.25, 1.25], '--', c='k')
ax.plot(bounds, [0.80, 0.80], '--', c='k')

ax.text(bounds[1], 1.25, ' two clusters', va='center', fontdict={'size': 13})
ax.text(bounds[1], 0.80, ' three clusters', va='center', fontdict={'size': 13})
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
```

```python
linkage_array = average(X)
# 现在为包含簇之间距离的linkage_array绘制树状图
dendrogram(linkage_array)

# 在树中标记划分成两个簇或三个簇的位置
ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds, [3.25, 3.25], '--', c='k')
ax.plot(bounds, [1.90, 1.90], '--', c='k')

ax.text(bounds[1], 3.25, ' two clusters', va='center', fontdict={'size': 13})
ax.text(bounds[1], 1.9, ' three clusters', va='center', fontdict={'size': 13})
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
```

```python
linkage_array = complete(X)
# 现在为包含簇之间距离的linkage_array绘制树状图
dendrogram(linkage_array)

# 在树中标记划分成两个簇或三个簇的位置
ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds, [6, 6], '--', c='k')
ax.plot(bounds, [3.80, 3.80], '--', c='k')

ax.text(bounds[1], 6, ' two clusters', va='center', fontdict={'size': 13})
ax.text(bounds[1], 3.80, ' three clusters', va='center', fontdict={'size': 13})
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
```
